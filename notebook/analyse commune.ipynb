```python
filename = "communes-france-2025.csv"
file_path = os.path.join(os.getcwd(), "..", "data", filename)
df_ori = pd.read_csv(file_path, sep=",")
# On affiche les 5 premières lignes pour vérifier que tout s'est bien déroulé 
pd.set_option('display.max_columns', 50)
df_ori.head(5)
```

```python
# Anam]lyse des mixed types
for col in df_ori.columns:
    if df_ori[col].dtype == 'object':
        num_unique_values = df_ori[col].nunique()
        num_total_values = len(df_ori[col])
        if num_unique_values / num_total_values < 0.05:
            print(f"Colonne '{col}' a des mixed types. Nombre de valeurs uniques: {num_unique_values}, Total: {num_total_values}")
            print(df_ori[col].unique())
```
